## 5] Checking the outliers in the dataset and removing them

```python
# Boxplots gives outliers present in our dataframe

sns.boxplot(x = df['HP'])     
```

```python
sns.boxplot(x = df['Price'])
```

### As we can see in the Boxplot there are outliers in the dataset which can be removed by finding the IQR and using the 1.5IQR rule

```python
# Finding the Interquantile Range(IQR)

q1 = df_new.quantile(0.25)
q3 = df_new.quantile(0.75)
IQR = q3-q1
IQR
```

```python
# Implementing 1.5*IQR rule to remove the outliers.

df_new = df_new[~((df_new < (q1 - 1.5 * IQR)) |(df_new > (q3 + 1.5 * IQR))).any(axis=1)]
df_new.shape
```

## 6.1] To Find the top 10 Car Brands in our dataset 

```python
# Finfing the top ten car brands in our dataframe

counts = df_new['Make'].value_counts()*100/sum(df_new['Make'].value_counts())

# Top 10 car brands in our datset
popular_labels = counts.index[:10]
    
#Visualiztion
plt.figure(figsize=(10,5))
plt.barh(popular_labels, width=counts[:10])
plt.title('Top 10 Car brands')
plt.show()
```

## 6.2] To Find the Average Price of Top Ten Car Brands

```python
# Finding the average price of cars brands

Average_prices = df_new[['Make','Price']].loc[(df_new['Make'] == 'Volkswagen')|
               (df_new['Make'] == 'Suzuki')|
               (df_new['Make'] == 'Nissan')|
               (df_new['Make'] == 'Chevrolet')|
               (df_new['Make'] == 'Toyota')|
               (df_new['Make'] == 'Ford')|
               (df_new['Make'] == 'Dodge')|
               (df_new['Make'] == 'GMC')|
               (df_new['Make'] == 'Mazda')|
               (df_new['Make'] == 'Honda')].groupby('Make').mean()
Average_prices
```
